### Solve
$$y'''' - \sin(y) + 2xy = 0$$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def dSdt(S,x):
    y1, y2, y3, y4 = S
    return np.array([
        y2,
        y3,
        y4,
        np.sin(y1) - 2*x*y1
    ])
```

```python
x = np.linspace(0,100,3600)
S0 = np.array([0,1,0,-1])
answer = odeint(dSdt,y0=S0,t=x)
```

```python
y = answer.T[0]
y_d = answer[:,1]
y_dd = answer[:,2]
y_ddd = answer[:,3]
```

```python
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
