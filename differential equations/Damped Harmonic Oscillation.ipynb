## Damped Harmonic Oscillation
$$m\ddot{x} + \beta\dot{x} + kx = 0$$
* $ \beta = $ damping factor
* $ k = $ spring constant

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def dSdt(S,t,m,b,k):
    x1 , x2 = S
    return [
        x2,
        -1/(m) * (b * x2 + k * x1)
    ]
```

```python
S0 = np.array([0,4])
t = np.linspace(0.01,20,1000)
m = 0.02
b = 0.01
k = 0.1
solution = odeint(dSdt,y0=S0,t=t,args=(m,b,k))
```

```python
position = solution[:,0]
velocity = solution[:,1]
```

```python
plt.plot(t,position,label="Position")
plt.plot(t,velocity,label="Velocity")
plt.legend()
plt.xlabel(r"$t\longrightarrow$")
plt.ylabel('Position and Velocity')
plt.title('Damped Harmonic Oscillator')
plt.grid()
```

### Total Energy

```python
T = (1/2)*m*velocity*velocity
V = (1/2)*k*position*position
E = T + V
```

```python
plt.plot(t,V)
plt.plot(t,T)
```
