## Solve
$$\frac{dy^2}{dx^2} + y\big(\frac{dy}{dx}\big)^2 + e^x = 0$$

## Solution

$\sf let,$
* $y_1 = y   $
* $y_2 = \frac{dy_1}{dx} $


$$ \begin{array}{c}
\frac{dy_1}{dx} = y_2 \\
\frac{dy_2}{dx} = -y_1y_2^2 - e^x \\
\end{array} $$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def dSdt(S,x):
    y1, y2 = S
    return np.array([
        y2,
        -y1*y2**2 - np.exp(x)
    ])
```

```python
x = np.linspace(0,100,3600)
S0 = np.array([1,0])
answer = odeint(dSdt,y0=S0,t=x)
```

```python
y = answer.T[0]
y_dashed = answer.T[1]
```

```python
plt.plot(x,y,label=r"$y'$")
plt.plot(x,y_dashed,label=r"$y''$")
plt.xlabel('x')
plt.ylabel('Function Values')
plt.title('Solutions to the Differential Equation')
plt.grid(True)
plt.legend()
plt.show()
```
