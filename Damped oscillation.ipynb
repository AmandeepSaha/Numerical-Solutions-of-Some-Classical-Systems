## Solve this ODE

$$\frac{d^2y}{dx^2} + 4\frac{dy}{dx} + 5y = 0$$
initial condition $y(0) = 1$ and $y'(0) = 0$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def dSdx(S,x):
    y1, y2 = S
    dy1dx = y2
    dy2dx = -0.5*y2 - 2*y1
    return [
        dy1dx,
        dy2dx
    ]
```

```python
S0 = [1,0]
x = np.linspace(-10,10,1000)
```

```python
solution = odeint(dSdx,y0=S0,t=x)
```

```python
plt.plot(x,solution.T[0],label="position")
plt.plot(x,solution.T[1],label="velocity")
plt.legend()
```
